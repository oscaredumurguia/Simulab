```python
def es_primo(numero):
    """Verifica si un número es primo."""
    if numero < 2:
        return False
    for i in range(2, int(numero**0.5) + 1):
        if numero % i == 0:
            return False
    return True

def primos_anteriores(limite):
    """Devuelve una lista de todos los números primos menores que un límite dado."""
    primos = []
    for i in range(2, limite):
        if es_primo(i):
            primos.append(i)
    return primos

# Aquí pedimos el número al usuario
numero_limite = int(input("Ingresa un número: "))

lista_primos = primos_anteriores(numero_limite)
print(f"Los números primos anteriores a {numero_limite} son: {lista_primos}")

```
