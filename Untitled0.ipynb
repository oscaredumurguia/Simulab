####*Toxicidad en la atmosfera de las nanopartículas*####
En el artículo del año 2024 sobre la toxicidad en el ire de nanopartículas hablas sobre la existencia de nanopartículas dañinas en el aire que se crean en diferentes procesos de los cuales no se han analizado lo suficiente, estas nanopartículas la mayoría de veces se generan por emisiones de coches, industria entre otros, y también se pueden generar con fuegos y erupciones volcanicas además de microplásticos.

Para ser considerados nanopartículas deben de tener un tamaño menor de 100 nm, la mayoría de estas son de carbonos, o metales pesados las cuales se encuentran presentes en areas urbanas afectando al 99% de la población, estas son altamente citotoxicas haciendo daño a las células y dañado por esto el tejido, afectando corazón, cerebro y pulmones, estan tienen diferentes morfologías, es un tema que tiene poco tiempo de investigado y debe de ser  analizado para ver como se puede prevenir esta aglomeraciones o contrarestalas.
